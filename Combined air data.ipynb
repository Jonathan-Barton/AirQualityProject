```python
import pandas as pd
import scipy as sc
from scipy.stats import norm

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (20, 6)
plt.style.use('ggplot')
```

```python
ozonecleaned = pd.read_csv(
    'OzoneMaytoDec.csv', 
    index_col=['Date Local', 'State Name', 'County Name']
)

ozonecleaned = ozonecleaned[['Arithmetic Mean ppb', 'AQI']].groupby(ozonecleaned.index).mean()

no2cleaned = pd.read_csv(
    'NO22020MaytoDec.csv', 
    index_col=['Date Local', 'State Name', 'County Name']
)

no2cleaned = no2cleaned[['Arithmetic Mean cleaned', 'AQI']].groupby(no2cleaned.index).mean()

pm25cleaned = pd.read_csv(
    'pm25MaytoDec.csv', 
    index_col=['Date Local', 'State Name', 'County Name']
)

pm25cleaned = pm25cleaned[['Arithmetic Mean', 'AQI']].groupby(pm25cleaned.index).mean()
```
